```python
import pandas as pd
import matplotlib.pyplot as plt
people=['Vaishnavi','Varun','Anusha','Pralekha','Manideep']
age=[20,19,18,17,16]
height=[3,4,5,6,7]
weight=[40,50,60,70,80]
data={
    'Name':people,
    'Age':age,
    'Height':height,
    'Weight':weight
}
df=pd.DataFrame(data)
print(df) #Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df['Age'],df['Weight'],color='blue',marker='o')
plt.title('Scatter plot of Age vs Weight')
plt.xlabel('Age(years)')
plt.ylabel('Weight(kg)')
#Histogram
plt.figure(figsize=(8,6))
plt.hist(df['Age'],bins=5,color='orange',edgecolor='black')
plt.title("Histogram of Age Distribution")
plt.xlabel('Age(years)')
#Bar plot
plt.figure(figsize=(8,6))
plt.bar(df['Name'],df['Height'],color='Green')
plt.title('Bar char of individuals Height')
plt.xlabel('Name')
plt.ylabel('Height(cm)')
plt.show()
```
